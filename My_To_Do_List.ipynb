```python
class ToDoList:
    def __init__(self):
        self.tasks = []

    def add_task(self, title, description=""):
        """Menambahkan tugas baru"""
        self.tasks.append({
            "title": title,
            "description": description,
            "completed": False
        })

    def remove_task(self, title):
        """Menghapus tugas"""
        self.tasks = [task for task in self.tasks if task["title"] != title]

    def mark_completed(self, title):
        """Menandai tugas sebagai selesai"""
        for task in self.tasks:
            if task["title"] == title:
                task["completed"] = True
                break

    def display_tasks(self):
        """Menampilkan semua tugas"""
        if not self.tasks:
            print("Tidak ada tugas.")
        else:
            for task in self.tasks:
                status = "✔" if task["completed"] else "✖"
                print(f"{status} {task['title']} - {task['description']}")

# Antarmuka pengguna sederhana berbasis teks

def menu():
    todo = ToDoList()
    while True:
        print("\n=== Menu ToDoList ===")
        print("1. Tampilkan semua tugas")
        print("2. Tambah tugas")
        print("3. Tandai tugas selesai")
        print("4. Hapus tugas")
        print("5. Keluar")

        pilihan = input("Pilih menu (1-5): ")

        if pilihan == "1":
            todo.display_tasks()
        elif pilihan == "2":
            title = input("Judul tugas: ")
            description = input("Deskripsi (opsional): ")
            todo.add_task(title, description)
            print("Tugas ditambahkan.")
        elif pilihan == "3":
            title = input("Judul tugas yang selesai: ")
            todo.mark_completed(title)
            print("Tugas ditandai selesai.")
        elif pilihan == "4":
            title = input("Judul tugas yang dihapus: ")
            todo.remove_task(title)
            print("Tugas dihapus.")
        elif pilihan == "5":
            print("Terima kasih!")
            break
        else:
            print("Pilihan tidak valid. Coba lagi.")

menu()
```
