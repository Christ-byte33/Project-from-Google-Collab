```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
import random
from IPython.display import display
import tkinter as tk
from tkinter import filedialog
```

```python
# Upload file
def upload_file():
    from google.colab import files  # hanya untuk Google Colab
    uploaded = files.upload()
    file_name = next(iter(uploaded))
    print(f"File dipilih: {file_name}")
    return file_name
```

```python
# Sebelum memproses data
def preprocess_data(file_path):
    df = pd.read_excel(file_path) # Membaca file
    print("\nDATA AWAL:")
    display(df.head())

    df.replace('?', np.nan, inplace=True) # Mengganti semua tanda '?' menjadi NaN
    for col in df.columns:
        if df[col].dtype == 'object':
            df[col].fillna(df[col].mode()[0], inplace=True)
            df[col] = LabelEncoder().fit_transform(df[col])
        else:
            df[col].fillna(df[col].mean(), inplace=True)

    return df
```

```python
# Normalisasi data
def normalize_data(df):
    X = df.values
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    return X_scaled
```

```python
# Kmeans
def euclidean(a, b):
    return np.sqrt(np.sum((a - b) ** 2))

def kmeans_custom(X, k, max_iters=100):
    n_samples, n_features = X.shape
    centroids = X[random.sample(range(n_samples), k)]
    for _ in range(max_iters):
        clusters = [[] for _ in range(k)]
        for idx, point in enumerate(X):
            distances = [euclidean(point, centroid) for centroid in centroids]
            cluster_idx = np.argmin(distances)
            clusters[cluster_idx].append(idx)

        new_centroids = np.array([np.mean(X[cluster], axis=0) if cluster else centroids[i]
                                  for i, cluster in enumerate(clusters)])
        if np.allclose(centroids, new_centroids):
            break
        centroids = new_centroids

    labels = np.zeros(n_samples)
    for cluster_idx, cluster in enumerate(clusters):
        for idx in cluster:
            labels[idx] = cluster_idx
    return labels, centroids
```
